```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
cancer = load_breast_cancer()
```

```python
data = pd.DataFrame(cancer.data, columns=cancer.feature_names)
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data_scaled = scaler.fit_transform(data)
data_target = scaler.fit_transform(cancer.data)
```

```python
import numpy as np
data['radius_category'] = pd.cut(data['mean radius'], bins=[0,10,20, np.inf],labels=['0-10','10-20','>30'])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data_scaled, data_target, test_size=0.3, train_size=0.7, shuffle=True)
```
