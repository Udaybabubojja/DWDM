```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['target'] = iris.target
```

```python
from sklearn.model_selection import train_test_split
y=data['target']
x=data.drop(columns='target')
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,train_size=0.8, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=42)
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
import numpy as np
new_data = np.array([[5.0,3,5,1.3]])
new_pred = model.predict(new_data)
new_pred
```

```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/race_results_20240101_20240131.csv')
data.head()
```

```python
mean_value = data['horse_win_value'].mean()
data['horse_win_value'].fillna(mean_value, inplace=True)
data.isnull().sum()
```

```python
data.drop(columns=['horse_margin', 'horse_late_start'], inplace=True)
data
```

```python
y=data['horse_rate']
data.drop(columns='horse_rate', inplace=True)
```

```python

x = data.iloc[:, [2,3,12,18,19,20]]
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, train_size=0.7, shuffle=True)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
